```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
cleandata_pd = pd.read_csv(csv_path, encoding = 'iso-8859-1', low_memory=False)
cleandata_pd.head()
```

```python
# Inspect all columns
cleandata_pd.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_coders_pd = cleandata_pd.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_coders_pd.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
reduced_coders_pd= reduced_coders_pd.replace({0.0: "No", 1.0: "Yes"})
reduced_coders_pd.head()
```

```python
# Calculate total number of respondents in survey
total_surveyed = len(reduced_coders_pd.index)
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_bootcamp = reduced_coders_pd.loc[reduced_coders_pd["AttendedBootcamp"] == "Yes"]
attended_bootcamp.count()
```

```python
# Calculate average age of attendees
average_age = attended_bootcamp["Age"].mean()
average_age

# Calculate how many people attended a bootcamp
people_attended = attended_bootcamp["AttendedBootcamp"].count()
people_attended
# Calculate how many attendees hold degrees
hold_degrees = attended_bootcamp["SchoolDegree"].count()
hold_degrees
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
total_gender = attended_bootcamp["Gender"].count()
male_gender = attended_bootcamp["Gender"].value_counts()['male']
female_gender = attended_bootcamp["Gender"].value_counts()['female']
non_gender = total_gender - male_gender - female_gender

# Calculate percentage of respondents who attended a bootcamp
percent_attended = people_attended / total_surveyed * 100

# Calculate percentage of respondents belonging to each gender
male_percent = (male_gender / total_gender)* 100
female_percent = (female_gender / total_gender)* 100
non_gender_percent = (non_gender / total_gender)* 100

# Calculate percentage of attendees with a school degree
percent_degree = hold_degrees / total_surveyed * 100

# Calculate average post-bootcamp salary
avg_postbootcamp_salary = attended_bootcamp["BootcampPostSalary"].mean()
```

```python
# Create a new table consolodating above calculations
bootcamp_summary = pd.DataFrame({"Total Surveyed": [total_surveyed],
                                "Total Bootcamp Attendees": [people_attended],
                                "Percent Attended": [percent_attended],
                                "Average Age": [average_age],
                                "Percent Male": [male_percent],
                                "Percent Female": [female_percent],
                                "Percent Non Gender Specific": [non_gender_percent],
                                "Holds A Degree": [hold_degrees],
                                "Average Post Bootcamp Salary": [avg_postbootcamp_salary]
                            })
bootcamp_summary = bootcamp_summary[["Total Surveyed", "Total Bootcamp Attendees", "Percent Attended", "Average Age", "Percent Male",
                                   "Percent Female", "Percent Non Gender Specific", "Holds A Degree", "Average Post Bootcamp Salary"]]
bootcamp_summary = bootcamp_summary.round(2)

bootcamp_summary
```

```python
# Improve formatting before outputting spreadsheet
bootcamp_summary["Percent Male"] =bootcamp_summary["Percent Male"].map("{0:,.2f}%".format)
bootcamp_summary["Percent Female"] =bootcamp_summary["Percent Female"].map("{0:,.2f}%".format)
bootcamp_summary["Percent Attended"] =bootcamp_summary["Percent Attended"].map("{0:,.2f}%".format)
bootcamp_summary["Percent Non Gender Specific"] =bootcamp_summary["Percent Non Gender Specific"].map("{0:,.2f}%".format)
bootcamp_summary["Holds A Degree"] =bootcamp_summary["Holds A Degree"].map("{0:,.2f}%".format)
bootcamp_summary["Average Post Bootcamp Salary"] =bootcamp_summary["Average Post Bootcamp Salary"].map("{0:,.2f}%".format)
bootcamp_summary
```

```python
# Export to Excel
bootcamp_summary.to_excel("output/Bootcamppart1.xlsx", index=False)
```
